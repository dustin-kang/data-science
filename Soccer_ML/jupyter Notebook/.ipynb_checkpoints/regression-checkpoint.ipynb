# Spliting Dataset
- target Í∞íÏùÑ `value_eur` ÏúºÎ°ú ÌïòÍ≥† ÏÑ†ÏàòÎì§Ïùò Îä•Î†•ÏùÑ ÌÜµÌï¥ ÏÑ†ÏàòÎì§Ïùò Í∞ÄÏπòÎ•º ÏòàÏ∏°Ìï† ÏòàÏ†ï.

```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt

# Ï†ÑÏ≤¥ Ïª¨Îüº Ï∂úÎ†•ÌïòÍ∏∞
pd.set_option('display.max_columns', None)

warnings.filterwarnings("ignore")

df = pd.read_csv('../data/reg_data.csv')

df.drop(columns=['position'], axis=1, inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split 

# ÌÉÄÍ≤ü(target)
target = 'value_eur'


# ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ ÏÖãÍ≥º ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ ÏÖã Î∂ÑÎ¶¨
train, test = train_test_split(df, random_state=2, train_size=.75)
# ÌõàÎû∏ Îç∞Ïù¥ÌÑ∞ÏÖãÏóêÏÑú ÌõàÎ†®Îç∞Ïù¥ÌÑ∞ ÏÖãÍ≥º Í≤ÄÏ¶ù Îç∞Ïù¥ÌÑ∞ ÏÖã Î∂ÑÎ¶¨
train, val = train_test_split(train, random_state=2, train_size=.75)

# ÌäπÏÑ±(features)
features = train.columns.drop(target)
```

```python
train.shape, val.shape, test.shape
```

```python
# X (features)
X_train = train[features] 
X_val = val[features]
X_test = test[features]

# y (Target)
y_train = train[target]
y_val = val[target] 
y_test = test[target]
```

## Regression Modeling

```python
# ÌååÏù¥ÌîÑÎùºÏù∏ #
from sklearn.pipeline import make_pipeline, Pipeline

# Ïù∏ÏΩîÎçî #
from category_encoders import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler

# Î™®Îç∏ÎßÅ #
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

# ÌèâÍ∞Ä #
from sklearn.metrics import mean_absolute_error,r2_score, mean_squared_error
```

### make pipeline
- `str` Îç∞Ïù¥ÌÑ∞ ÏûêÎ£åÌòïÏúºÎ°ú Ïù¥Î£®Ïñ¥ÏßÑ Ïª¨ÎüºÏùÑ 0Í≥º 1Î°ú Î∞îÍæ∏Îäî **OneHotEncoder(ÏõêÌï´ Ïù∏ÏΩîÎçî)**Î•º ÏÇ¨Ïö©Ìï®.
- Ïù¥ÌõÑ ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ÏÖãÏùÄ `fit_transform`ÏùÑ ÏßÑÌñâÌï®.
- Í≤ÄÏ¶ù, ÏãúÌóò Îç∞Ïù¥ÌÑ∞ÏÖãÏùÄ `transform`ÏùÑ ÏßÑÌñâÌï®.


```python
pipe = make_pipeline(
    OneHotEncoder(cols=['preferred_foot','body_type','work_rate_Attaking','work_rate_defensive'], use_cat_names=True),
    # SimpleImputer(missing_values=np.nan, strategy='mean'),
    MinMaxScaler()
)

X_train_transform = pipe.fit_transform(X_train)
X_val_transform = pipe.transform(X_val)
X_test_transform = pipe.transform(X_test)
```

### Linear Regression (ÏÑ†ÌòïÌöåÍ∑Ä)

```python
reg = LinearRegression()
reg = reg.fit(X_train_transform, y_train)
y_pred = reg.predict(X_val_transform)
```

```python
print(f"- MAE : {mean_absolute_error(y_val, y_pred)}")
print(f"- RMSE : {mean_squared_error(y_val, y_pred)**0.5}")
print(f"- R2Score : {r2_score(y_val, y_pred)}")
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=[15,10])
sns.regplot(x=y_val, 
           y=y_pred, 
           fit_reg=True);

plt.title("Linear Regression Plot(Value_EUR)", fontsize=20)
plt.xlabel('Truth', fontsize=14)

plt.ylabel('Prediction', fontsize=14)
plt.show()
```

### Ridge

```python
reg = Ridge()
reg = reg.fit(X_train_transform, y_train)
y_pred = reg.predict(X_val_transform)

print(f"MAE : {mean_absolute_error(y_val, y_pred)}")
print(f"RMSE : {mean_squared_error(y_val, y_pred)**0.5}")
print(f"R2Score : {r2_score(y_val, y_pred)}")
```

```python
fig = plt.figure(figsize=[15,10])
sns.regplot(x=y_val, 
           y=y_pred, 
           fit_reg=True);

plt.title("Ridge Plot(Value_EUR)", fontsize=20)
plt.xlabel('Truth', fontsize=14)

plt.ylabel('Prediction', fontsize=14)
plt.show()
```

### RandomForest (ÎûúÎç§ Ìè¨Î†àÏä§Ìä∏)

```python
reg = RandomForestRegressor()
reg = reg.fit(X_train_transform, y_train)
y_pred = reg.predict(X_val_transform)

print(f"MAE : {mean_absolute_error(y_val, y_pred)}")
print(f"RMSE : {mean_squared_error(y_val, y_pred)**0.5}")
print(f"R2Score : {r2_score(y_val, y_pred)}")
```

### XGBoost

```python
reg = XGBRegressor()
reg = reg.fit(X_train_transform, y_train)
y_pred = reg.predict(X_val_transform)

print(f"MAE : {mean_absolute_error(y_val, y_pred)}")
print(f"RMSE : {mean_squared_error(y_val, y_pred)**0.5}")
print(f"R2Score : {r2_score(y_val, y_pred)}")

"""
[23:20:11] WARNING: /workspace/src/objective/regression_obj.cu:152: reg:linear is now deprecated in favor of reg:squarederror.
MAE : 427093.0390904744
RMSE : 1251733.4461512708
R2Score : 0.949373346492528
"""
```
